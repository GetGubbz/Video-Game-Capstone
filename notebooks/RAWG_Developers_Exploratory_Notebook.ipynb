```python
import requests
import json
import pandas as pd
import os
```

```python
with open('keys.json') as fi:
    credentials = json.load(fi)
api_key = credentials["developer_url"]
```

```python
response = requests.get(api_key)
response
```

```python
res = response.json()
```

```python
res.keys()
```

```python
developers = pd.DataFrame(res['results'])
developers.head(2)
```

```python
developers['games'][0]
```

```python
developers['name'].value_counts()
```

```python
l = []
for i in range(10):
    df = pd.json_normalize(developers['games'][i])
    df['company'] = developers['name'][i]
    l.append(df)

developer_df = pd.concat(l).reset_index()
developer_df
```
